```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/water_dataX.csv', encoding="ISO-8859-1")
```

```python
data.fillna(0, inplace=True)
```

```python
data['Temp'] = pd.to_numeric(data['Temp'], errors = 'coerce')
data['D.O. (mg/l)'] = pd.to_numeric(data['D.O. (mg/l)'], errors = 'coerce')
data['PH'] = pd.to_numeric(data['PH'], errors='coerce')
data['B.O.D. (mg/l)'] = pd.to_numeric(data['B.O.D. (mg/l)'], errors='coerce')
data['CONDUCTIVITY (µmhos/cm)'] = pd.to_numeric(data['CONDUCTIVITY (µmhos/cm)'], errors='coerce')
data['NITRATENAN N+ NITRITENANN (mg/l)'] = pd.to_numeric(data['NITRATENAN N+ NITRITENANN (mg/l)'], errors='coerce')
data['TOTAL COLIFORM (MPN/100ml)Mean'] = pd.to_numeric(data['TOTAL COLIFORM (MPN/100ml)Mean'], errors='coerce')
```

```python
data.dtypes
```

```python
start = 1
end = 1779
station = data.iloc[start:end, 0]
location = data.iloc[start:end ,1]
state = data.iloc[start:end, 2]
do = data.iloc[start:end, 4].astype(np.float64)

value=0

ph = data.iloc[ start:end, 5]
co = data.iloc [start:end, 6].astype(np.float64)

year = data.iloc[start:end, 11]
tc = data.iloc[2:end, 10].astype(np.float64)


bod = data.iloc[start:end, 7].astype(np.float64)
na = data.iloc[start:end, 8].astype(np.float64)
na.dtype
```

```python
data.head()
```

```python
data = pd.concat([station,location,state,do,ph,co,bod,na,tc,year], axis=1)

data.columns = ['station','location','state','do','ph','co','bod','na','tc','year']
```

```python
data.head()
```

```python
#calulation of Ph
data['npH']=data.ph.apply(lambda x: (100 if (8.5>=x>=7)
                                 else(80 if  (8.6>=x>=8.5) or (6.9>=x>=6.8)
                                      else(60 if (8.8>=x>=8.6) or (6.8>=x>=6.7)
                                          else(40 if (9>=x>=8.8) or (6.7>=x>=6.5)
                                              else 0)))))
```

```python
#calculation of dissolved oxygen
data['ndo']=data.do.apply(lambda x:(100 if (x>=6)
                                 else(80 if  (6>=x>=5.1)
                                      else(60 if (5>=x>=4.1)
                                          else(40 if (4>=x>=3)
                                              else 0)))))
```

```python
#calculation of total coliform
data['nco']=data.tc.apply(lambda x:(100 if (5>=x>=0)
                                 else(80 if  (50>=x>=5)
                                      else(60 if (500>=x>=50)
                                          else(40 if (10000>=x>=500)
                                              else 0)))))
```

```python
#calc of B.D.O
data['nbdo']=data.bod.apply(lambda x:(100 if (3>=x>=0)
                                 else(80 if  (6>=x>=3)
                                      else(60 if (80>=x>=6)
                                          else(40 if (125>=x>=80)
                                              else 0)))))
```

```python
#calculation of electrical conductivity
data['nec']=data.co.apply(lambda x:(100 if (75>=x>=0)
                                 else(80 if  (150>=x>=75)
                                      else(60 if (225>=x>=150)
                                          else(40 if (300>=x>=225)
                                              else 0)))))
```

```python
#Calulation of nitrate
data['nna']=data.na.apply(lambda x:(100 if (20>=x>=0)
                                 else(80 if  (50>=x>=20)
                                      else(60 if (100>=x>=50)
                                          else(40 if (200>=x>=100)
                                              else 0)))))
```

```python
data.head()
```

```python
data['wph']=data.npH * 0.165
data['wdo']=data.ndo * 0.281
data['wbdo']=data.nbdo * 0.234
data['wec']=data.nec* 0.009
data['wna']=data.nna * 0.028
data['wco']=data.nco * 0.281
```

```python
data['wqi']=data.wph+data.wdo+data.wbdo+data.wec+data.wna+data.wco
```

```python
data.head()
```

```python
# Convert columns to numeric
numeric_cols = ['year','do', 'ph', 'co', 'bod', 'na', 'tc', 'npH',	'ndo',	'nco','nbdo', 'nec', 'nna', 'wph', 'wdo', 'wbdo', 'wec', 'wna', 'wco', 'wqi']
data[numeric_cols] = data[numeric_cols].apply(pd.to_numeric, errors='coerce')

print(data[numeric_cols])
# Now you can perform aggregation
aggregated_data = data[numeric_cols].groupby('year').mean()
ag = aggregated_data
```

```python
data = ag.reset_index(level = 0, inplace = False)
data.head()
```

```python
def print_water_quality(wqi):

    if wqi >= 95 and wqi <= 100:
        quality = "Excellent"
        description = "Water quality is protected with a virtual absence of threat. The conditions are very close to natural levels"
    elif wqi >= 80 and wqi <= 94:
        quality = "Good"
        description = "Water quality is protected with a minor degree of threat. The conditions rarely depart from natural levels"
    elif wqi >= 65 and wqi <= 79:
        quality = "Fair"
        description = "Water quality is usually protected but occasionally threatened. The conditions sometimes depart from natural levels"
    elif wqi >= 45 and wqi <= 64:
        quality = "Poor (Marginal)"
        description = "Water quality is frequently threatened. The conditions often depart from natural levels"
    elif wqi >= 0 and wqi <= 44:
        quality = "Very Poor (Poor)"
        description = "Water quality is almost always threatened. The conditions usually depart from natural levels"
    else:
        quality = "Invalid WQI value"
        description = "Please enter a value between 0 and 100."

    print(f"Water Quality: {quality}")
    print(f"Description: {description}")

# Example usage
print_water_quality(92)
```

```python
def predict_water_class(total_coliform, ph, dissolved_oxygen, biochemical_oxygen_demand, free_ammonia=None, electrical_conductivity=None, sodium_absorption_ratio=None, boron=None):
    if (total_coliform <= 50 and
        6.5 <= ph <= 8.5 and
        dissolved_oxygen >= 6 and
        biochemical_oxygen_demand <= 5):
        water_class = "A"
        designated_use = "Drinking Water Source without conventional treatment but after disinfection"
    elif (total_coliform <= 500 and
          ph >= 6.5 and
          dissolved_oxygen >= 5 and
          biochemical_oxygen_demand <= 20):
        water_class = "B"
        designated_use = "Outdoor bathing (Organised)"
    elif (total_coliform <= 5000 and
          6 <= ph <= 9 and
          dissolved_oxygen >= 4 and
          biochemical_oxygen_demand <= 20):
        water_class = "C"
        designated_use = "Drinking water source after conventional treatment and disinfection"
    elif (6.5 <= ph <= 8.5 and
          dissolved_oxygen >= 4 and
          free_ammonia is not None and free_ammonia <= 1.2):
        water_class = "D"
        designated_use = "Propagation of Wild life and Fisheries"
    elif (6.0 <= ph <= 8.5 and
          electrical_conductivity is not None and electrical_conductivity <= 2250 and
          sodium_absorption_ratio is not None and sodium_absorption_ratio <= 26 and
          boron is not None and boron <= 2):
        water_class = "E"
        designated_use = "Irrigation, Industrial Cooling, Controlled Waste disposal"

    # Condition for domestic use
    if (6.5 <= ph and ph <= 8.5 and
        dissolved_oxygen >= 4 and
        total_coliform <= 50):
        print( "Domestic Use")
        print("Water suitable for domestic purposes")

    return water_class, designated_use
```

```python
# Example usage
total_coliform = 40
ph = 7.2
dissolved_oxygen = 6.5
biochemical_oxygen_demand = 4
water_class, designated_use = predict_water_class(total_coliform, ph, dissolved_oxygen, biochemical_oxygen_demand)
print(f"Water Class: {water_class}")
print(f"Designated-Best-Use: {designated_use}")
```
